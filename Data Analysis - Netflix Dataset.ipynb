## DATA VISUALIZATION 

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (12,8))
sns.countplot(x = 'Rating' ,data = data)
```

```python
plt.figure(figsize = (35,6))
sns.countplot(x = 'Year',data = data)
```

```python
#distribution according to the rating
data['Rating'].value_counts().plot.pie(figsize = (20,35))
```

```python
data['Country'].value_counts()[:10].plot.pie(figsize = (30,45))
```

```python
plt.figure(figsize = (14,8))
data[data['Category'] == 'Movie']['Year'].value_counts().plot(kind="bar",color="Red")
plt.title("Frequency of Movies which were released in different years and are available on netflix")
```

```python
plt.figure(figsize = (14,8))
data[data['Category'] == 'TV Show']['Year'].value_counts().plot(kind = "bar",color = "Blue")
plt.title("Frequency of TV Shows which were released in different years and are available on netflix")
```

```python
plt.figure(figsize = (12,6))
data[data['Category'] == 'Movie']['Type'].value_counts()[:10].plot(kind = 'barh',color = 'Black')
plt.title("Top 10 Genres of Movies")
```

```python
plt.figure(figsize = (12,6))
data[data['Category'] == 'TV Show']['Type'].value_counts()[:10].plot(kind = 'barh',color = 'Yellow')
plt.title("Top 10 Genres of TV Show")
```
